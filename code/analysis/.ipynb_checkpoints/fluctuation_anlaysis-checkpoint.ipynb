## Define analysis functions

```python
def sigma_GN(df,n_bins):
    X,Y = df.bf.values,df.gfp.values
    bin_ends = np.linspace(X[0],X[-1],n_bins+1)
    # inds = [np.array([k for k in range(len(X)) if X[k]> bin_ends[j] and X[k] < bin_ends[j+1]])\
     #       for j in range(n_bins)]
    inds = [np.where((X > bin_ends[j]) & (X <= bin_ends[j+1])) for j in range(n_bins)]
    Xb = (bin_ends[1:]+bin_ends[:-1])/2
    Y_var = np.array([np.var(Y[i]) for i in inds])
    return Xb,Y_var

def sigma_GN_theory(N,p):
    if p>0.5:
        return p/(2*p-1)*N**(1+np.log(1-p)/np.log(2))
    else:
        return p/(1-2*p)*N**(2+2*np.log(1-p)/np.log(2))
```

# Illustrate phase transition on constant probability model

```python
f = lambda gt,t,n,cell_type: np.random.normal(60,0.3*60) # kernel
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution

n_bins = 50
fig,ax = plt.subplots(figsize=(10,5))

p = 0.2
model = sims.constant_probability_model(f,f0,p)
data = model.run_ensemble(300,200,tmax=500,dt_sample=5)
X,Y_var = sigma_GN(data,n_bins)
ax.plot(X,Y_var,'C0o',label='p = {}'.format(p),fillstyle='none')
ax.plot(X,[sigma_GN_theory(x,p) for x in X],'C0-')

p = 0.4
model = sims.constant_probability_model(f,f0,p)
data = model.run_ensemble(300,200,tmax=500,dt_sample=5)
X,Y_var = sigma_GN(data,n_bins)
ax.plot(X,Y_var,'C1o',label='p = {}'.format(p),fillstyle='none')
ax.plot(X,[sigma_GN_theory(x,p) for x in X],'C1-')

p = 0.6
model = sims.constant_probability_model(f,f0,p)
data = model.run_ensemble(300,200,tmax=500,dt_sample=5)
X,Y_var = sigma_GN(data,n_bins)
ax.semilogy(X,Y_var,'C2o',label='p = {}'.format(p),fillstyle='none')
ax.plot(X,[sigma_GN_theory(x,p) for x in X],'C2-')

ax.legend(frameon=False)
plt.xlabel(r'$N$');
plt.ylabel(r'${\rm var}(G|N)$');
plt.savefig(FIG_PATH+"phase_transition_constant_rate.pdf",bbox_inches = "tight")
```

## effects of carrying capcity

```python
def f(gt,t,n,cell_type):
    if n>=120:
        tau = 10000000
    else:
        tau = 60/(1-n/120)
    return np.random.normal(tau,0.3*tau) # kernel
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution
```

# Replicate data for broken state model

```python
# define various functions used in simulations
def f(gt,t,n,cell_type):
    if cell_type==1:
        return abs(np.random.normal(200,0.3*100)) # time to be broken
        #return np.random.exponential(60)
    else:
        return abs(np.random.normal(30,0.3*30)) # time to double for non-broken cells
    
    
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution
p = 0.3 # the probability to break
Q = lambda gt,t: np.array([
    [1-p,p,0],
    [0,0,1],
    [0,0,1]
])
types_names = ['m','br','gfp']
model = sims.bellman_harris_model_base(f,f0,Q,types_names)
```
