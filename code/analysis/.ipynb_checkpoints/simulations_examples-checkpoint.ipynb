
## Example: constant probability model

To build a constant probability model, we would do the following:

```python
# define various functions used in simulations
f = lambda gt,t,n,cell_type: np.random.normal(60,0.3*60) 
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution
p = 0.3 # the probability to turn green
Q = lambda gt,t: np.array([
    [1-p,p],
    [0,1]
])
types_names = ['m','gfp']
```

We can then build our model and simulate a single well

```python
importlib.reload(sims)
model = sims.bellman_harris_model_base(f,f0,Q,types_names)
output = model.run_well(500,tmax=500)
output.head()
```

```python
fig,ax = plt.subplots()
ax.plot(output.time,output.gfp)
ax.semilogy(output.time,output.bf)
ax.set_ylabel('cells')
ax.set_xlabel('time')
```

We can also generate an ensemble of wells to mimic the experiments:

```python
data = model.run_ensemble(100,1000,dt_sample=0.3)
```

## Example: carrying capcity model

Since $f$ may be dependent on the population size we can incorperate a carrying capacity

```python
def f(gt,t,n,cell_type):
    if n>=120:
        tau = 10000000
    else:
        tau = 60/(1-n/120)
    return np.random.normal(tau,0.3*tau) # kernel
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution
model = sims.constant_probability_model(f,f0,0.1)
```

```python
data = model.run_ensemble(10,200,dt_sample=1,tmax=5000)
```

```python
plt.plot(data.groupby(['time']).mean().bf.values)
```

# More complex models

## Example: broken non-growing state model

```python
# define various functions used in simulations
def f(gt,t,n,cell_type):
    if cell_type==1:
        return abs(np.random.normal(100,0.3*100)) # time to be broken
    else:
        return abs(np.random.normal(30,0.3*30)) # time to double for non-broken cells
    
    
f0 = lambda : np.random.normal(60,0.3*60) # lag time distribution
p = 0.3 # the probability to break
Q = lambda gt,t: np.array([
    [1-p,p,0],
    [0,0,1],
    [0,0,1]
])
types_names = ['m','br','gfp']
importlib.reload(sims)
model = sims.bellman_harris_model_base(f,f0,Q,types_names)
```

```python
data = model.run_ensemble(10,200,dt_sample=1,tmax=5000)
```

```python
plt.plot(data.groupby(['time']).mean().br)
plt.plot(data.groupby(['time']).mean().m)
plt.semilogy(data.groupby(['time']).mean().gfp)
```
