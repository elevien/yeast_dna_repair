# Define useful functions

```python
class bellman_harris_model:
    def __init__(self,f,f0,p):
        '''
        Simulate exponentially growing population

        Input:
            Nmax        - the maximum number of cells to generate
            seed        - generation time of ancestral cell
                          (relevant if there are correlations)
        '''
        self.f = f
        self.f0 = f0
        self.p = p
  
    
    def run_well(self,Nmax,*,dt_sample=0.01,tmax=100):
        '''
        Simulate exponentially growing population

        Input:
            Nmax        - the maximum number of cells to generate
            seed        - generation time of ancestral cell
                            (relevant if there are correlations)

        Optional Input:
            tmax        - the maximum time to run
            dt_sample   - frequancy to save samples of the number of cells


        Output:
            N          - the number of cells at sample times
            T          - sample times
            L          - estimate of population growth rate from fitting
        '''

        gt0 = self.f0()
        cells_gt = np.zeros(Nmax) # generation times
        cells_dt = np.zeros(Nmax) # division times
        cells_type = np.zeros(Nmax,dtype=int) # type of cell (0 if no cell in slot)
        cells_gt[0] = gt0
        cells_dt[0] = gt0
        cells_type[0] = 1
        
        N = [1]
        M = [0]
        n=1
        T = [0.]
        t_sample = 0.
        n = 1
        m = 0
        t = 0.
        r = np.random.rand()
#         if r<0.5:
#             self.p = 0.1
#         else:
#             self.p = 0.5

        while n<Nmax and t<tmax:
            ind = np.argmin(cells_dt[0:n])
            mother_dt = cells_dt[ind]
            mother_gt = cells_gt[ind]
            mother_type =  cells_type[ind]
            t_next = mother_dt
            gt1 = self.f(mother_gt,n)
            gt2 = self.f(mother_gt,n)
            cells_gt[ind] = gt1
            cells_dt[ind] = gt1+t_next
            cells_gt[n] = gt2
            cells_dt[n] = gt2+t_next
            
            t = t_next
            # save samples
            t_last = T[-1]
 
                
            while t-t_last>dt_sample:
                t_last += dt_sample
                T.append(t_last)
                N.append(n)
                M.append(m)
                t_sample = 0.
           

            # decide color of daughters
            if mother_type==1:
                r = np.random.rand()
                if r < self.p:#0.1*t/tmax:
                    cells_type[n] = 2
                    m = m + 1
                else:
                    cells_type[n] = 1
                    
                r = np.random.rand()
                if r < self.p: #0.1*t/tmax:
                    cells_type[ind] = 2
                    m = m + 1
                else:
                    cells_type[ind] = 1
                    
            else:
                cells_type[n] = 2
                cells_type[ind] = 2
                m = m + 1
            n = n + 1
                    
         
          
        return np.array(T),np.array(N),np.array(M)
    
    def run_ensemble(self,Nwells,Nmax,**kwargs):
        T = []
        N = []
        M = []
        for k in range(Nwells):
            t,n,m = self.run_well(Nmax,**kwargs)
            T.append(t)
            N.append(n)
            M.append(m)
    
        min_length = np.min([len(t) for t in T])
        T = np.concatenate([t[:min_length] for t in T])
        N = np.concatenate([n[:min_length] for n in N])
        M = np.concatenate([m[:min_length] for m in M])
        wells = np.concatenate([k*np.ones(min_length,dtype=int) for k in range(Nwells)])
        return pd.DataFrame({"time":T,"bf":N,"gfp":M,"well":wells})
```

```python
def binned(X,Y,bin_size):
    """
    group the X values into bins of size bin_size and
    compute the average and standard deviation of the
    Y values
    """
    Y_sorted = [x for _,x in sorted(zip(X,Y))]
    X_sorted = sorted(X)
    L = len(X)
    X_chunks = [X_sorted[x:x+bin_size] for x in range(0,L-1,bin_size)]
    Y_chunks = [Y_sorted[x:x+bin_size] for x in range(0,L-1,bin_size)]
    Xb = np.array([np.mean(x) for x in X_chunks])
    Yvar = np.array([np.var(x) for x in Y_chunks])
    Yb = np.array([np.mean(x) for x in Y_chunks])
    return Xb,Yb,Yvar
```

```python
def binned(X,Y,bin_size):
    """
    group the X values into bins of size bin_size and
    compute the average and standard deviation of the
    Y values
    """
    Y_sorted = [x for _,x in sorted(zip(X,Y))]
    X_sorted = sorted(X)
    L = len(X)
    X_chunks = [X_sorted[x:x+bin_size] for x in range(0,L-1,bin_size)]
    Y_chunks = [Y_sorted[x:x+bin_size] for x in range(0,L-1,bin_size)]
    Xb = np.array([np.mean(x) for x in X_chunks])
    Yvar = np.array([np.var(y) for y in Y_chunks])
    Yb = np.array([np.mean(y) for y in Y_chunks])
    return Xb,Yb,Yvar
```

```python
def sigma_GN(df,n_bins):
    X,Y = df.bf.values,df.gfp.values
    bin_ends = np.linspace(X[0],X[-1],n_bins+1)
    # inds = [np.array([k for k in range(len(X)) if X[k]> bin_ends[j] and X[k] < bin_ends[j+1]])\
     #       for j in range(n_bins)]
    inds = [np.where((X > bin_ends[j]) & (X <= bin_ends[j+1])) for j in range(n_bins)]
    Xb = (bin_ends[1:]+bin_ends[:-1])/2
    Y_var = np.array([np.var(Y[i]) for i in inds])
    return Xb,Y_var
```
